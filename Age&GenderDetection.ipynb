# **Age and Gender Detection**

Importing these libraries to view image

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

**Facial Attribute Analysis**
Deepface also comes with a strong facial attribute analysis module including age, gender, facial expression (including angry, fear, neutral, sad, disgust, happy and surprise) and race (including asian, white, middle eastern, indian, latino and black) predictions. Here we are predicting age and gender only

We will be using these models:

opencv, ssd, dlib

```python
backends = [
  'opencv',
  'ssd',
  'dlib',
  'mtcnn'
]
```

**Function Definition:** This is the function that gets the image path and runs the image through these models to get age and gender from these. Age is dicretized using the if-block code. The group having the highest frequency is returned.

```python
def get_gad(img_path):
  age_arr=[]
  face = False
  for model in backends:
    try:

      demographies = DeepFace.analyze(img_path = img_path,
              detector_backend = model,
              actions = ['age', 'gender']
        )
      face = True
      age=demographies[0]['age']
      age=str(age)
      if(age>='13' and age <='17'):
        age='13-17'
      if(age>='18' and age <='24'):
        age='18-24'
      if(age>='25' and age <='34'):
        age='25-34'
      if(age>='35' and age <='44'):
        age='35-44'
      if(age>='45' and age <='54'):
        age='45-54'
      if(age>='55' ):
        age='55+'
      age_arr.append(age)
      gender=demographies[0]['dominant_gender']
    except Exception as ex:
      print(ex)

  if face == True:
    age_counts = {}
    for age_group in age_arr:
        if age_group in age_counts:
            age_counts[age_group] += 1
        else:
            age_counts[age_group] = 1
    most_common_age_group = max(age_counts, key=age_counts.get)
    print('\n')
    print('*******Results********')
    print('The detected age group is: ', most_common_age_group)
    print('The detected gender is: ', gender)
  else:
    print('No face detected')
```
